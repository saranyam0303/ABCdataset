# Importing Libraries and Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("dataset.csv")
df
```

# The "height" column by replacing it with random numbers between 150 and 180

```python
df["Height"]=np.random.randint(150, 180,size=len(df))
df
```

# bar graph of height of employees

```python
br=df["Height"].value_counts()
plt.figure(figsize=(9,5))
br.plot.bar( color='g') 
plt.title("Height of employees") 
plt.xlabel("Height") 
plt.ylabel("Number of employees") 
  
plt.show() 
```

# 1.Distribution of employees across each team and the percentage split relative to the total number of employees.


```python
#1. Determine the distribution of employees across each team and
#calculate the percentage split relative to the total number of employees.


team_dist = df['Team'].value_counts()
total_employees = len(df)
team_per = (team_dist / total_employees) * 100
print("Distribution of employees across each team:",team_dist)
print("\n Percentage split relative to the total number of employees:",team_per)
```

# Bar plot distribution of employees across each team

```python
br=df["Team"].value_counts()
plt.figure(figsize=(9,5))
br.plot.bar( color='b') 
plt.title("No. of Employees Across Each Team") 
plt.xlabel("Team") 
plt.ylabel("Number of employees") 
  
plt.show() 
```

# 2. Segregate employees based on their positions within the company.

```python
position_dis = df['Position'].value_counts()

print("Position Distribution:", position_dis)
```

# Graphical Representation of employees and position

```python


plt.figure(figsize=(5, 5))
sns.countplot(x='Position', data=df,width=.50)
plt.title('Employees and positions')
plt.xlabel('Position')
plt.ylabel('Number of Employees')
plt.show()
```

# 3. Identify the predominant age group among employees

```python
Age_bins = [15 ,20 ,25 ,30 ,35 ,40 ,45 , 50]
Age_label = ['15-19','20-25','26-29', '30-35','36-39','40-45', '45-50']
df['Age_Group'] = pd.cut(df['Age'], bins=Age_bins, labels=Age_label, right=False)
age_group_distribution = df['Age_Group'].value_counts().sort_index()
print("Age Group Distribution:",age_group_distribution)
```

# Graphical Representation of age group of employees


```python
#Graphical Representation of age group of employees

plt.figure(figsize=(5, 5))
sns.countplot(y='Age_Group', data=df,width=.50,palette = "Set2")
plt.title('Employees and Age group')
plt.xlabel('Ages')
plt.ylabel('Number of Employees')
plt.show()
```

# 4. Discover which team and position have the highest salary expenditure

```python
team = df.groupby('Team')['Salary'].sum()
position= df.groupby('Position')['Salary'].sum()
max_team_salary = team.idxmax()
max_position_salary = position.idxmax()

print("Team with Highest Salary Expenditure:", max_team_salary)
print("Position with Highest Salary Expenditure:", max_position_salary)
```

# Graphical Representation of Total Salary Expenditure by Team


```python
plt.figure(figsize=(10, 6))
team.plot(kind='bar', color='y')
plt.title('Total Salary Expenditure by Team')
plt.xlabel('Team')
plt.ylabel('Total Salary ($)')
plt.show()
```

# Graphical representation of Total Salary Expenditure by Team

```python
plt.figure(figsize=(10, 6))
position.plot(kind='bar', color='b')
plt.title('Total Salary Expenditure by Team')
plt.xlabel('Team')
plt.ylabel('Total Salary ($)')
plt.show()
```
